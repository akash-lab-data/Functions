1. What is the difference between a function and a method in Python?
- A function in Python is a block of reusable code, which perform a specific function.
- It is defined using the `def`  keyword.
- Function is independent and can be called directy.
- Example:

```python
def greet(name):
    return f"Hello, {name}!"

print(greet("Akash"))
```

Here, greet() is a function that take a argument return a greeting.

- **Method:**
- Method is also a type of function, but it is associated with object.
- Method are defined inside classes and called using object of the class.
- Method always has its first perametre as *`self`*  and also refer the object itself.
- Example:



```python
class Person:
  def __init__(self, name):
    self.name = name


  def greet(self):   #method
     return f"Hello,{self.name}!"

p1 = Person("Akash")
print(p1.greet())      #method calling
```

Here, greet() is a method of a class `Person`





2. Explain the concept of function arguments and parameters in Python.
- **Parameter:**
- Parameters are variables that are defined in the defination of function and present under the parentheses.
- They act like placeholder which receives value, when function is called.
- Example:

```python
def greet(name):
  print(f"Hello {name}!")
```

- **Arguments:**
- Arguments are actual value you pass to the function when you call it.
- They fill in the parameters.
- Example:

```python
greet("Akash") # "Akash" is the argument.
```

âœ… Types of Arguments in Python:
- Python supports different way to pass argument.
- **Positional Arguments:**
- Value are assigned to parameter in order.


```python
def add(a, b):
  return a + b

print(add(8, 6))   #a = 8, b= 6
```

- **Keyword Arguments:**
- specify parameter name while passing the value.

```python
def student(name, age):
  print(f"Name: {name}, Age: {age}")

student(age = 22, name = "Akash")
```

- **Default Arguments:**
- Parameter have default value if no parameter is provided.


```python
def greet(name = "Guest"):
  print(f"Hello, {name}!")

greet()         # uses default â†’ Guest
greet("Akash")  # overrides deafult
```

- **Arbitrary Arguments:**
- `*args`â†’ allows multiple positional arguments.
- `**kwargs`â†’ allows multiple keyword arguments.
- Example:



```python
def add_all(*args):
  return sum(args)


print(add_all(10, 20, 30, 40))
```

```python
def show_info(**kwargs):
  for key, value in kwargs.items():
    print(f"{key}, {value}")


show_info(name = "Akash", age = 22, city = "Jhajjar")
```

6. Explain the concept of generators in Python and how they are defined.
- Genertor is special functions that are return an iterator by using the `yield` keyword.
- Unlike normal function, they cannot return all values at once - instead they generate value lazily(one at a time).
- Example:

```python
def count_up_to(n):
  i = 1
  while i <= n:
   yield i
   i += 1

for num in count_up_to(5):
  print(num)
```

7. What are the advantages of using generators over regular functions?
- **Memory efficient:** Don't store all items in memory, generates them when needed.
- **Lazy Evaluation**: Useful for handling infinite or very large sequences.

10.  What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

| Function                                    | Purpose                                                          | Returns             |
| ------------------------------------------- | ---------------------------------------------------------------- | ------------------- |
| `map(func, iterable)`                       | Applies `func` to each element                                   | Iterator of results |
| `filter(func, iterable)`                    | Keeps elements where `func(x)` is True                           | Filtered iterator   |
| `reduce(func, iterable)` (from `functools`) | Applies `func` cumulatively to reduce iterable to a single value | Single value        |


- Example:

```python
from functools import reduce

nums = [1, 2, 3, 4, 5]

print(list(map(lambda x : x*x,nums)))
print(list(filter(lambda x : x%2 == 0 , nums)))
print(reduce(lambda x,y : x+y, nums))
```

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(n):
  if n < 2:
    return False
  for i in range(2, int(n**0.5) + 1):
    if n % i == 0:
      return False
  return True

prime = [x for x in range(1, 201) if is_prime(x)]
print(prime)
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
#   terms.
def fib(n):
  a = 0
  b = 2
  for i in range(n): #if n = 5 0, 1, 2, 3, 4,
      yield a
      a, b = b, a+b


for n in fib(200):
    print(n, end=" ")   #iterator class created here for the specific number 20


```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power_of_two(exp):
  for i in range(exp):
    yield 2**i

print(list(power_of_two(10)))
```

```python
#7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_file(filename):
  with open(filename, "r") as f:
    for line in f:
      yield line

#Example:  assuming test.text exist
for line in read_file("test.text"):
  print(line)
```
